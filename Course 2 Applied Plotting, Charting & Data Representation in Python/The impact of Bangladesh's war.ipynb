```python
df = df2.copy()
year = 15
df = df.iloc[:, 5:20]         #16 = 1975 ; 26 = 1985 ; 36 = 1995 ; 45 = 2005 ; 56 = 2015 ;
war = 6
labelSize = 18

p11 = 'Industry (including construction), value added (constant 2010 US$)'
p12 = 'Manufacturing, value added (constant 2010 US$)'
p21 = 'Exports of goods and services (current US$)'
p22 = 'Imports of goods and services (current US$)'
p31 = 'GDP per capita (constant 2010 US$)'
p41 = 'Gross national expenditure (constant LCU)'
p42 = 'Gross domestic income (constant LCU)'

#fig = plt.figure()
fig = plt.figure (figsize=(18,10))
fig.suptitle('The impact of Bangladesh\'s War on 1971', size = 22)


ax1 = plt.subplot(2, 2, 1)
plt.plot(df.loc[p31], label = 'GDP per capita')
plt.legend(loc="upper left")
plt.ylabel("US$", size = labelSize)
plt.xticks(np.arange(0, year, step=1))
plt.axvline(x=war, color="darkred")
ax1.text(.48,.75, "War", size=15, ha="center", transform=ax1.transAxes)


ax2 = plt.subplot(2, 2, 2)
plt.plot(df.loc[p11], label = 'Industry value added')
plt.plot(df.loc[p12], label = 'Manufacturing value added')
#plt.plot(df.loc[p3], label = 'Imports of goods')

plt.axvline(x=war, color="darkred")
plt.legend(loc="upper left")
plt.ylabel("US$ in Billion", size = labelSize)
plt.xticks(np.arange(0, year, step=1))
ax2.text(.48,.75, "War", size=15, ha="center", transform=ax2.transAxes)
#axes.set_xlim([xmin,xmax])
#ax.set_ylim([70,100])


ax3 = plt.subplot(2, 2, 3)
plt.plot(df.loc[p21], label = 'Exports of goods')
plt.plot(df.loc[p22], label = 'Import of goods')
plt.legend(loc="upper left")
plt.ylabel("US$ in Billion", size = labelSize)
plt.xticks(np.arange(0, year, step=1))
plt.axvline(x=war, color="darkred")
ax3.text(.48,.57, "War", size=15, ha="center", transform=ax3.transAxes)



ax4 = plt.subplot(2, 2, 4)
plt.plot(df.loc[p41], label = 'Gross National expenditure')
plt.plot(df.loc[p42], label = 'Gross domestic income')
plt.legend(loc="upper left")
plt.ylabel('constant LCU', size = labelSize)
plt.xticks(np.arange(0, year, step=1))
plt.axvline(x=war, color="darkred")
ax4.text(.48,.57, "War", size=15, ha="center", transform=ax4.transAxes)


ax1.set_title('Impact on GDP', size=18)
ax2.set_title('Impact on Business', size=18)
ax3.set_title('Impact on Import Export', size=18)
ax4.set_title('Impact on Income and Expenditure', size=18)
```
